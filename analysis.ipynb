## Function for linking fixture difficulty

```python
# FUNCTION
import requests
url = "https://fantasy.premierleague.com/api/bootstrap-static/"

response = requests.get(url)
team_df = pd.DataFrame(response.json()['teams']).set_index('id').drop(columns=[
       'code', 'draw', 'form',
       # 'id',
       'loss', 'name', 'played', 'points',
       'position', 'short_name',
       # 'strength', 
       'team_division', 'unavailable',
       'win',
       #'strength_overall_home', 'strength_overall_away',
       #'strength_attack_home', 'strength_attack_away', 'strength_defence_home',
       #'strength_defence_away',
       'pulse_id'])
```

```python
oppenent_home = team_df.loc[history['opponent_team'],['strength', 'strength_overall_home',
       'strength_attack_home', 'strength_defence_home']]
oppenent_home.columns = oppenent_home.columns.map(lambda x : 'opponent_' + '_'.join(x.split('_')[0:2]))
len(oppenent_home)
```

```python

oppenent_away = team_df.loc[history['opponent_team'],['strength', 'strength_overall_away',
       'strength_attack_away', 'strength_defence_away']]
oppenent_away.columns = oppenent_away.columns.map(lambda x : 'opponent_' + '_'.join(x.split('_')[0:2]))
len(oppenent_away)
```

## Concat the fixuture difficulty

```python
#history['opponent_strength','opponent_strength_overall','opponent_strength_attack','opponent_strength_defence'] =
fixture_difficulty = oppenent_home.where(history['was_home'], oppenent_away).reset_index().drop(columns=['id'])
history = pd.concat([history,fixture_difficulty], axis=1)
```

```python
history[list(opponent_dummies.columns)] = opponent_dummies
history = history.drop(['element','fixture','kickoff_time','opponent_team'], axis=1)
history
```

```python
history['goals_for'] = history['team_h_score'].where(history['was_home'], history['team_a_score'])
history['goals_against'] = history['team_a_score'].where(history['was_home'], history['team_h_score'])
history = history.drop(columns=['team_h_score','team_a_score'])
```

```python
columns_to_drop = ['value', 'transfers_balance', 'selected', 'transfers_in','transfers_out','opponent_strength', 'opponent_strength_overall',
       'opponent_strength_attack', 'opponent_strength_defence']
new_history = pd.concat([history.groupby('round').sum().drop(columns=columns_to_drop), history.groupby('round')[columns_to_drop].mean()], axis=1)
new_history
```

```python
x=[np.arange(i-9,i+1) for i in range(10,len(new_history)+1)]
x
#x= np.arange(0,10)
```
