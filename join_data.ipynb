```python
new_understat = pd.DataFrame.from_dict(understat).drop(columns=['id', 'time', 'goals', 'assists', 'yellow_cards', 'red_cards', 'position'])
new_understat.head()
```

```python
names_ud = new_understat['player_name']
names_ud.head()
```

```python
names = new_player_data[['first_name', 'second_name', 'web_name']].agg(' '.join, axis=1)
names.head()
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
fuzz.partial_ratio("Mo Salah", "Mohammad Salah")
```

```python
best_match = {}
confidence = {}
correct_matches = {}
for player in names:
    player = ' '.join(sorted(set(player.split())))
    ratings = {fuzz.token_sort_ratio(player, un_player):un_player for un_player in names_ud}
    best_match[player] = ratings[max(ratings)]
    confidence[player] = max(ratings)
    if max(ratings) > 90:
        correct_matches[player] = ratings[max(ratings)]
```
