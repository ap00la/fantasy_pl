```python
web_names = new_player_data['web_name']
web_names.head()
```

```python
names =pd.concat([full_names, web_names], axis=1).rename({0:'name', 'web_name':'short_name'}, axis='columns')
names
```

```python
best_match = {}
confidence = {}
correct_matches = {}
for player in names:
    player = ' '.join(sorted(set(player.split())))
    # Players with multiple names
    ratings = {fuzz.token_sort_ratio(un_player, player):un_player for un_player in names_ud if len(un_player.split()) > 1}
    
    
    ratings = {fuzz.token_sort_ratio(un_player, player):un_player for un_player in names_ud if len(un_player.split()) > 1}
    best_match[player] = ratings[max(ratings)]
    confidence[player] = max(ratings)
    if max(ratings) > 90:
        correct_matches[player] = ratings[max(ratings)]
```

```python
best_match = {}
confidence = {}
correct_matches = {}
for un_player in names_ud:
    if len(un_player.split()) == 1:
        ratings = {fuzz.token_sort_ratio(un_player, name):name for name in web_names}
    else:
        ratings = {fuzz.token_sort_ratio(un_player, name):name for name in names}
    
    best_match[un_player] = ratings[max(ratings)]
    confidence[un_player] = max(ratings)
    
    if max(ratings) > 90:
        correct_matches[player] = ratings[max(ratings)]
```
