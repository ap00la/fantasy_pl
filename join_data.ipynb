```python
columns_to_drop = ['code', 'cost_change_event', 'cost_change_event_fall', 'cost_change_start', 'cost_change_start_fall',
                   'ep_next', 'ep_this', 'photo', 'special', 'squad_number', 'team_code', 'transfers_in_event',
                   'transfers_out_event', 'influence_rank_type', 'creativity_rank_type', 'threat_rank_type',
                   'ict_index_rank_type', 'corners_and_indirect_freekicks_text', 'direct_freekicks_text', 'penalties_text']
new_player_data = pd.DataFrame.from_dict(player_data, orient='index').drop(columns=columns_to_drop)
new_player_data.head()
```

```python
new_understat = pd.DataFrame.from_dict(understat).drop(columns=['id', 'time', 'goals', 'assists', 'yellow_cards', 'red_cards', 'position', 'team_title'])

# Rename required players
rename_players = {'Rúben Dias':'Rúben Santos Gato Alves Dias'}
new_understat['player_name'] = new_understat['player_name'].replace(rename_players)
new_understat
```

```python
names_ud = new_understat['player_name']
names_ud.head()
```

```python
player_with_mins = new_player_data[new_player_data.minutes> 0]
player_with_mins.head()
```

```python
full_names = player_with_mins[['first_name', 'second_name']].agg(' '.join, axis=1)
web_names = player_with_mins['web_name']
web_names.head()
```

```python
names =pd.concat([full_names, web_names], axis=1).rename({0:'name', 'web_name':'short_name'}, axis='columns')
names
```

```python
from fuzzywuzzy import fuzz
import copy
```

```python
un_players = list(copy.deepcopy(names_ud))
names_del = copy.deepcopy(names)
conf_score = 100
best_match = {}
confidence = {}
print(type(un_players))
print(type(names_del))
while True:

    for un_player in un_players:
        if len(un_player.split()) > 1 and conf_score > 70:
            # Players with multiple names
            ratings = {fuzz.partial_ratio(un_player, player[1][0]):player[1][0] for player in names_del.iterrows()}
        else:

            # Player with one name
            # direct match tried as well as a fuzzy match. Direct match given confidence score of 150
            ratings = {150 if un_player == player[1][1] else fuzz.partial_ratio(un_player, player[1][1]):player[1][0] for player in names_del.iterrows()}
        
        if max(ratings) >= conf_score:
            best_match[un_player] = ratings[max(ratings)]
            confidence[un_player] = max(ratings)
            un_players.remove(un_player)
            names_del.drop(names_del.index[names_del['name'] == ratings[max(ratings)]], inplace=True)

    conf_score -= 2
    if not len(un_players):
        break
       
```

```python
dump = [best_match, confidence]
```

```python
name_change_understat = copy.deepcopy(new_understat)
name_change_understat

# Rename required players
name_change_understat['player_name'] = name_change_understat['player_name'].replace(best_match)
name_change_understat
```

```python
df_to_join = player_with_mins.join(names['name']).drop(columns=['first_name','second_name','web_name']).rename(columns={'name':'player_name'})
df_to_join
```

```python
df = pd.merge(df_to_join.reset_index(), name_change_understat, how ='inner', on=['player_name', 'player_name'])
df
```
