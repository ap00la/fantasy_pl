```python
import pandas as pd
import numpy as np
import import_tool
import datetime
import pre_process as pp
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
today = '_'.join(str(datetime.datetime.today()).split(' ')[0].split('-'))

weeks = 3
filename = '../.data/' + today + '_data_ML_' + str(weeks) + '_week.csv'

# Initially try to load the csv file, ready for ML
try:
    df = pd.read_csv(filename)

except FileNotFoundError:
    
    # If CSV ready for ML not present
    # Try to load the gw_data
    try:
        gw_df = pd.read_pickle("../.data/" + today + "_gw_df")
    
    # If todays data is not found then re-download the data
    except FileNotFoundError:
        _, gw_df = pp.main(save_to_file=True)
    
    df = pp.create_ml_df(gw_df, weeks=weeks)
    df.to_csv(filename,index=False)
```

```python
df2 = df[df['target'] != 0]
```

```python
df_train, df_test =train_test_split(df)
df_train
```

```python
X_train = df_train.drop(columns=['target','id','element_type','player_name'])
y_train = df_train['target']

X_test = df_test.drop(columns=['target','id','element_type','player_name'])
y_test = df_test['target']
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model = RandomForestRegressor(max_depth=10, max_features=10)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

The results are not pretty!

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_true=y_test, y_pred=y_pred)
```
